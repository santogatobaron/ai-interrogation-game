```python
@tool
def evaluate_question_tool(input_data: dict) -> dict:
    """
    Evaluate the detective's question using the detective vector database.

    Produces signals used for:
    - interrogation progress
    - suspect behaviour
    - player feedback
    """

    question = input_data["question"]

    # retrieve similar interrogation questions (hide final parts of the interrogation)
    allowed_phases = list(range(0, state.phase + 3))
    allowed_phases = [p for p in allowed_phases if p >= 0]

    results = detective_vector_store.similarity_search_with_score(
        question,
        k=5,
        filter={"phase": {"$in": allowed_phases}}
    )

    print("\n================ VECTOR SEARCH DEBUG ================")
    print("Question:", question)
    print("Current phase:", state.phase)
    print("Allowed phases:", allowed_phases)

    retrieved_phases = []
    similarities = []
    docs = []

    for i, (doc, score) in enumerate(results):

        phase = doc.metadata.get("phase", None)
        text = doc.page_content.replace("\n", " ")

        print(f"\nResult {i+1}")
        print("raw score:", score)
        print("phase:", phase)
        print("text:", text[:200])

        # Pinecone normally returns similarity (cosine)
        similarity = score

        similarities.append(similarity)
        docs.append(doc)

        if phase is not None:
            retrieved_phases.append(int(phase))
        else:
            retrieved_phases.append(None)

    print("\nSimilarities used:", similarities)
    print("Retrieved phases:", retrieved_phases)
    print("=====================================================\n")

    # determine which phases are acceptable for conversational alignment
    valid_alignment_phases = list(range(0, state.phase + 2))

    aligned_similarities = [
        s for s, p in zip(similarities, retrieved_phases)
        if p in valid_alignment_phases
    ]

    if len(retrieved_phases) == 0:

        phase_alignment_ratio = 0
        mean_similarity = 0

    else:

        phase_alignment_ratio = len(aligned_similarities) / len(retrieved_phases)

        if aligned_similarities:
            mean_similarity = sum(aligned_similarities) / len(aligned_similarities)
        else:
            mean_similarity = 0

    print("\n=========== ALIGNMENT DEBUG ===========")
    print("Valid alignment phases:", valid_alignment_phases)
    print("Aligned similarities:", aligned_similarities)
    print("phase_alignment_ratio:", phase_alignment_ratio)
    print("mean_similarity:", mean_similarity)
    print("=======================================\n")

    # define phases allowed to generate interrogation progress
    progress_phases = [
        state.phase - 1,
        state.phase,
        state.phase + 1
    ]

    progress_phases = [p for p in progress_phases if p >= 0]

    print("\n=========== PROGRESS PHASE DEBUG ===========")
    print("Progress phases:", progress_phases)
    print("============================================\n")

    weighted_scores = []
    weighted_docs = []

    print("\n=========== PROGRESS SCORING ===========")

    for similarity, phase, doc in zip(similarities, retrieved_phases, docs):

        print("\nCandidate")
        print("phase:", phase)
        print("similarity:", similarity)

        if phase not in progress_phases:
            print("→ rejected (phase not allowed)")
            continue

        if phase == state.phase:
            weight = 1.0
        elif phase == state.phase + 1:
            weight = 0.7
        elif phase == state.phase - 1:
            weight = 0.6
        else:
            weight = 0

        weighted = similarity * weight

        print("weight:", weight)
        print("weighted score:", weighted)

        weighted_scores.append(weighted)
        weighted_docs.append(doc)

    print("\nweighted_scores:", weighted_scores)
    print("=========================================\n")

    if weighted_scores:
        max_similarity = max(weighted_scores)
        best_index = weighted_scores.index(max_similarity)
        best_doc = weighted_docs[best_index]
    else:
        max_similarity = 0
        best_doc = None

    progress_delta = max_similarity

    print("\n=========== FINAL PROGRESS DEBUG ===========")
    print("max_similarity:", max_similarity)
    print("progress_delta:", progress_delta)

    if best_doc:
        print("\nBEST MATCH CHUNK:")
        print(best_doc.page_content[:300])
        print("phase:", best_doc.metadata.get("phase"))

    print("============================================\n")

    return {
        "phase_alignment_ratio": phase_alignment_ratio,
        "mean_similarity": mean_similarity,
        "max_similarity": max_similarity,
        "progress_delta": progress_delta,
        "retrieved_phases": retrieved_phases
    }
```

```python
import pandas as pd

def debug_interrogation_questions(
    questions,
    session_id,
    phase=None,
    reset_progress=False
):
    """
    Debug multiple interrogation questions and display vector retrieval + game signals.
    
    Parameters
    ----------
    questions : list[str]
        Questions to test
    
    session_id : str
        Conversation session id
    
    phase : int | None
        Force the game phase before running
    
    reset_progress : bool
        Reset pressure / progress if desired
    """

    if phase is not None:
        print(f"\n🔧 Setting game phase → {phase}")
        state.phase = phase

    if reset_progress:
        print("🔧 Resetting pressure")
        state.pressure = 0

    rows = []

    print("\n====================================")
    print("DEBUG INTERROGATION RUN")
    print("Current phase:", state.phase)
    print("====================================")

    for q in questions:

        print("\n------------------------------------")
        print("Question:", q)
        print("------------------------------------")

        result = agent_with_memory.invoke(
            {"question": q},
            config={"configurable": {"session_id": session_id}}
        )

        state_update = result.get("state_update", {})

        row = {
            "question": q,
            "phase": state.phase,
            "pressure": state.pressure,
            "phase_alignment_ratio": result.get("phase_alignment_ratio"),
            "mean_similarity": result.get("mean_similarity"),
            "max_similarity": result.get("max_similarity"),
            "progress_delta": result.get("progress_delta"),
            "retrieved_phases": result.get("retrieved_phases"),
        }

        rows.append(row)

        print("Stephanie:", result["response"][:120], "...")

        if result.get("feedback"):
            print("Instinct:", result["feedback"])

    df = pd.DataFrame(rows)

    print("\n====================================")
    print("SUMMARY TABLE")
    print("====================================")

    display(df)

    return df
```

```python
def debug_retrieval(question, phase=None, k=5):

    print("\n==============================")
    print("QUESTION:")
    print(question)

    # retrieval
    results = detective_vector_store.similarity_search_with_score(
        question,
        k=k
    )

    retrieved_phases = []

    print("\n--- RETRIEVED CHUNKS ---")

    for i, (doc, score) in enumerate(results):

        phase_meta = doc.metadata.get("phase", "unknown")
        retrieved_phases.append(phase_meta)

        print(f"\nChunk {i+1}")
        print("score:", score)
        print("phase:", phase_meta)

        preview = doc.page_content[:200].replace("\n"," ")
        print("text:", preview)

    print("\n--- SUMMARY ---")

    print("retrieved_phases:", retrieved_phases)

    if phase is not None:

        aligned = sum(1 for p in retrieved_phases if p == phase)
        ratio = aligned / len(retrieved_phases)

        print("current_phase:", phase)
        print("alignment_ratio:", ratio)

    print("==============================\n")
```
